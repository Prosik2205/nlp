```python
import os
from pymongo import MongoClient
from pymongo.server_api import ServerApi
import gridfs

uri = "mongodb+srv://Sanya:Prosik@lecture.hc3ho.mongodb.net/?retryWrites=true&w=majority&appName=Lecture"
client = MongoClient(uri, server_api=ServerApi('1'))

try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)

    
db = client['Lecture']
fs = gridfs.GridFS(db)

def download_files(output_directory):
    for file_data in fs.find():
        file_path = os.path.join(output_directory, file_data.metadata['path'])
        os.makedirs(os.path.dirname(file_path), exist_ok=True)

        with open(file_path, 'wb') as file:
            file.write(file_data.read())
            print(f'File downloaded: {file_path}')

output_dir = r'C:\Users\prosi\Desktop\van' 
download_files(output_dir)
```
